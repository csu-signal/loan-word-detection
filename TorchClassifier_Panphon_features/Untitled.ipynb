```python
import pandas as pd
import unicodeblock.blocks
import numpy as np
```

```python
chinese_blocks = ['CJK_RADICALS_SUPPLEMENT', 'CJK_SYMBOLS_AND_PUNCTUATION', 'CJK_STROKES',\
                 'ENCLOSED_CJK_LETTERS_AND_MONTHS', 'CJK_COMPATIBILITY', 'CJK_UNIFIED_IDEOGRAPHS_EXTENSION_A',\
                 'CJK_UNIFIED_IDEOGRAPHS', 'CJK_COMPATIBILITY_IDEOGRAPHS', 'CJK_COMPATIBILITY_FORMS',\
                 'CJK_UNIFIED_IDEOGRAPHS_EXTENSION_B', 'CJK_UNIFIED_IDEOGRAPHS_EXTENSION_C',\
                 'CJK_UNIFIED_IDEOGRAPHS_EXTENSION_D', 'CJK_COMPATIBILITY_IDEOGRAPHS_SUPPLEMENT']
```

```python
filename = '../Datasets/Hard-Negatives/Chinese-English-Hard-Negatives.csv'
df = pd.read_csv(filename)
drop = []
for i in range(len(df)):
    if any([unicodeblock.blocks.of(c) in chinese_blocks for c in df.iloc[i]['loan_word_epitran']]):
        drop.append(i)

print(len(df))
df = df.drop(drop)
print(len(df))

df['original_unicode'] = "Latin"

df.to_csv(filename, index=False)
```
